## Histogramas

```python
df = pd.read_csv('schoolimprovement2010grants.txt')
```

```python
sns.distplot(df['Award_Amount'], kde=True, bins=20)

# al poner kde = False, quitamos la linea que sigue la distribucion de los datos
# el default es kde = True
plt.show()
```

## Personalizando graficos en Seaborn

```python
df = pd.read_csv('FY18_4050_FMRs.txt')
```

Podemos fijar el estilo por defecto de seaborn usando sns.set()

Para 'limpiar' el grafico, usamos plt.clf(), así los estilos de un grafico no interfieren con los demás

Podemos remover las lineas que se encuentran en los ejes de la siguiente manera

```python
sns.set_style('white')

sns.lmplot(data=df, x='pop2010', y='fmr_2')

# Remove the spines
sns.despine(top=True, right=True, left=True)

plt.show()
plt.clf()
```

### Colores

Tambien podemos crear nuestras propias secuencias de colores segun lo que necesitemos

```python
# creamos una paleta de 10 colores usando el sistema husl
sns.palplot(sns.color_palette('husl', 10))
plt.show()

# esta paleta se puede usar cuando no nos importa que los datos tengan un color por su orden o tamaño
```

```python
# paleta de 6 colores donde conforme nos acercamos al centro se vuelve mas claro
sns.palplot(sns.color_palette('coolwarm',9))
plt.show()

# usamos esto cuando los valores mas grandes y mas chicos son los que nos interesan
```

## Personalizando con matplotlib

Así como en el curso de matplotlib, tendremos que crear una figura y los ejes de la siguiente manera

```python
fig, ax = plt.subplots()

# notemos que debemos poner el eje en el que queremos nuestra figura
sns.distplot(df['fmr_3'], ax=ax)

# fijamos el nombre del eje x
ax.set(xlabel="3 Bedroom Fair Market Rent")

plt.show()
```

```python
fig, ax = plt.subplots()

sns.distplot(df['fmr_1'], ax=ax)

# fijamos el nombre del ejex, el maximo y el minimo valor que va a tomar x; por ultimos, ponemos titulo
ax.set(xlabel="1 Bedroom Fair Market Rent", xlim=(100,1500), title="US Rent")

plt.show()
```

```python
sns.set_palette(sns.color_palette('husl', 10))
fig, ax = plt.subplots()
sns.distplot(df['fmr_1'], ax=ax)

ax.set(xlabel="1 Bedroom Fair Market Rent", xlim=(100,1500), title="US Rent")

# Añadimos lineas verticales con la media y el promedio de los datos
ax.axvline(x=df['fmr_1'].median(), color='m', label='Median', linestyle='--', linewidth=2)
ax.axvline(x=df['fmr_1'].mean(), color='b', label='Mean', linestyle='-', linewidth=2)

# mostramos la leyenda de las lineas que hicimos arriba
ax.legend()
plt.show()
```

```python
# Creamos una figura con 1 columna y 2 columnas 
# ambos ejes, compartiendo el ejey de los datos
fig, (ax0, ax1) = plt.subplots(nrows=1, ncols=2, sharey=True)

# Plot the distribution of 1 bedroom apartments on ax0
sns.distplot(df['fmr_1'], ax=ax0)
ax0.set(xlabel="1 Bedroom Fair Market Rent", xlim=(100,1500))

# Plot the distribution of 2 bedroom apartments on ax1
sns.distplot(df['fmr_2'], ax=ax1)
ax1.set(xlabel="2 Bedroom Fair Market Rent", xlim=(100,1500))

plt.show()
```

Este nos ayuda a ver que tan 'precisa' fue la regresion lineal, lo ideal es que los datos se encuentren lo mas cerca de la linea que aparece

## Graficos usando matrices

### Grafico de calor (heatmap)

Este grafico nos ayuda para hacer un breve resumen de que tanto estan relacionadas dos variables. Para poder crear uno, necesitamos que los valores del dataframe sean solo numeros

```python
df = pd.read_csv('daily_show_guests_cleaned.txt')
```

```python
pd_crosstab = pd.crosstab(df["Group"], df["YEAR"])
print(pd_crosstab.head())

sns.heatmap(pd_crosstab)

# Rotate tick marks for visibility
plt.yticks(rotation=0)
plt.xticks(rotation=90)

plt.show()
```

Este grafico nos indica en que año y categoria se encuentran los valores mas grandes y pequeños

```python
pd_crosstab = pd.crosstab(df["Group"], df['YEAR'])

# Plot a heatmap of the table with no color bar and using the BuGn palette
sns.heatmap(pd_crosstab, cbar=True, cmap="BuGn", linewidths=0.3, annot=True)
# con los colores, lo menos es azul y lo mas es verde
# el parametro linewidths = 0.3 hace que los cuadrados no esten pegados del todo
# cbar = False hace que el cuadro de la leyenda no aparezca

plt.yticks(rotation=0)
plt.xticks(rotation=90)

plt.show()
plt.clf()
```

## PairGrid

Nos ayuda a ver si existe alguna relacion entre dos o mas variables, mostrando diversos graficos

```python
df = pd.read_csv('insurance_premiums.txt')
```

```python
g = sns.PairGrid(df, vars=["fatal_collisions", "premiums"])
g2 = g.map(plt.scatter)

plt.show()
plt.clf()
```

Podemos cambiar los gráficos que aparecen en las diagonales de la siguiente manera

```python
g = sns.PairGrid(df, vars=["fatal_collisions", "premiums"])
g2 = g.map_diag(plt.hist)
g3 = g2.map_offdiag(plt.scatter)

plt.show()
plt.clf()
```

## Pairplot

Es una manera simplificada de usar PairGrid

```python
# el ejercicio anterior lo podiamos hacer de la siguiente manera
sns.pairplot(data=df, vars=["fatal_collisions", "premiums"], kind='scatter')

plt.show()
plt.clf()
```

```python
sns.set_style('white')
sns.pairplot(data=df, vars=["fatal_collisions", "premiums"], kind='scatter', hue='Region', palette='RdBu',
             diag_kws={'alpha':.2})
# el parametro diag_kws = {'alpha':1} hace que el grafico tenga transparencia de 0.5

plt.show()
plt.clf()
```

Podemos cambiar los valores de los ejes, así, en vez de transponer los datos, se nos van a generar graficos diferentes entre si

```python
sns.pairplot(data=df, x_vars=["fatal_collisions_speeding", "fatal_collisions_alc"],
        y_vars=['premiums', 'insurance_losses'], kind='scatter', hue='Region',
        palette='husl')

plt.show()
plt.clf()
```

Podemos cambiar manualmente el tipo de graficos que queremos en la diagonal principal

```python
sns.pairplot(data=df, vars=["insurance_losses", "premiums"], kind='reg', palette='BrBG',
             diag_kind = 'kde', hue='Region')
# el parametro diag_kind es el que hace que cambie el tipo de grafico de la diagonal principal

plt.show()
plt.clf()
```

## JointGrid

Este metodo va a ser idoneo cuando queramos personalizar mas nuestra gráfica

```python
df = pd.read_csv('bike_share.txt')
```

```python
sns.set_style("whitegrid")
g = sns.JointGrid(x="hum", y="total_rentals", data=df, xlim=(0.1, 1.0)) 

# ponemos la regresion lineal en el centro y las distribuciones de los datos a la derecha y arriba
g.plot(sns.regplot, sns.distplot)

plt.show()
plt.clf()
```

## jointplot()

Podemos generar los mismo graficos que en jointgrid, pero de manera mas sencilla, limitando algunas cosas para personalizar

```python
sns.jointplot(x="hum", y="total_rentals", kind='reg', data=df)

plt.show()
plt.clf()
```

```python
sns.jointplot(x="temp", y="total_rentals", kind='reg', data=df, order=2, xlim=(0, 1))

plt.show()
plt.clf()
```

Revisamos que tan apropiado fue el modelo con un gráfico de residuo

```python
sns.jointplot(x="temp", y="total_rentals", kind='resid', data=df, order=2)

plt.show()
plt.clf()
```

```python
g = (sns.jointplot(x="temp", y="casual", kind='scatter', data=df, marginal_kws=dict(bins=10, rug=True))
    .plot_joint(sns.kdeplot))
    
plt.show()
plt.clf()
```

```python
g = (sns.jointplot(x="temp", y="registered", kind='scatter', data=df, marginal_kws=dict(bins=10, rug=True))
    .plot_joint(sns.kdeplot))

plt.show()
plt.clf()
```
